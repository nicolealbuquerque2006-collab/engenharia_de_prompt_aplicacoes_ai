```python

# Define uma lista de números inteiros
numeros = [1, 3, 6, 7, 8]

# Inicializa uma variável para armazenar a soma dos números
soma = 0

# Itera sobre cada número uma lista de 'numeros'
for num in numeros:
  soma += num

# Imprimir o valor final de todos os numeros
print(soma)
```
