```python
# Solicita os números ao usuário
num1_input = input("Digite o primeiro número: ")
num2_input = input("Digite o segundo número: ")

# Tenta converter os inputs para float
try:
    num1 = float(num1_input)
    num2 = float(num2_input)
except ValueError:
    print("Entrada inválida. Por favor, digite apenas números.")
else:
    # Solicita a operação ao usuário
    operation = input("Escolha a operação (+, -, *, /): ")

    result = None
    if operation == '+':
        result = add(num1, num2)
    elif operation == '-':
        result = subtract(num1, num2)
    elif operation == '*':
        result = multiply(num1, num2)
    elif operation == '/':
        result = divide(num1, num2)
    else:
        print("Operação inválida. Por favor, escolha entre +, -, * ou /.")

    if result is not None:
        print(f"O resultado da operação {num1} {operation} {num2} é: {result}")

Digite o primeiro número: 2
Digite o segundo número: 2
Escolha a operação (+, -, *, /): *
O resultado da operação 2.0 * 2.0 é: 4.0

def check_temperature_simulation(current_temp):
    """
    Simula a verificação de temperatura comparando uma temperatura fornecida
    pelo usuário com um limite fixo de 30°C.

    Args:
        current_temp (float): A temperatura atual informada pelo usuário.
    """
    fixed_threshold = 30 # Limite fixo conforme solicitado.

    print(f"Verificando a temperatura simulada {current_temp}°C com limite de {fixed_threshold}°C...")

    if current_temp > fixed_threshold:
        print(f"*** ALERTA: A temperatura simulada está superaquecendo! ({current_temp}°C > {fixed_threshold}°C) ***")
    else:
        print(f"A temperatura simulada ({current_temp}°C) está abaixo de {fixed_threshold}°C.")

# Solicita a temperatura atual ao usuário
current_temp_input = input("Digite a temperatura atual (em °C): ")

try:
    # Converte a entrada do usuário para um número
    current_temp = float(current_temp_input)

    # Executa a função de simulação de verificação de temperatura com o limite fixo
    check_temperature_simulation(current_temp)
except ValueError:
    print("Entrada inválida. Por favor, digite apenas um número para a temperatura.")

Digite a temperatura atual (em °C): 31
Verificando a temperatura simulada 31.0°C com limite de 30°C...
*** ALERTA: A temperatura simulada está superaquecendo! (31.0°C > 30°C) ***
Criar um programa que organize arquivos bagunçados em pastas automaticamente, baseando-se apenas na extensão do arquivo (ex: pdf, jpg, xls, doc, py, .ipynb.) use a biblioteca ose shutil. comente a execução duração o código use 10 arquivos de cada tipo
```
